En este notebook se muestra cómo usar la clase FiltroRubik, sus métodos y sus parámetros.  
Esta clase permite aplicar un filtro de Rubik a una image, además de robar paleta de colores e implementar a la imagen con un efecto pixelado.  

# Importar clase

```python
from filtrorubik import FiltroRubik
```

# Instanciar objeto

```python
filtro_rubik = FiltroRubik()
```

# Probando métodos

Probando el método filtro sólo con la imagen y el tamaño de los cuadrados pixelados.

```python
filtro_rubik.filtro("Einstein.jpg", 10) #sólo imagen y tamaño cuadrado
filtro_rubik.show()
```

Método filtro con la imagen, el tamaño de los cuadrados pixelados y una de las paleta de colores que ya tiene definida la clase.  
El nombre de las paletas son paleta1, paleta2 y paleta3.

```python
filtro_rubik.filtro('Einstein.jpg', 5, "paleta2") #probando con paleta integrada
filtro_rubik.show()
```

Al método se le puede pasar como argumento una paleta de colores personalizada. LA paleta debe ser una lista con los colores en cadena de texto en formato hexadecimal.

```python
paleta_personalizada = ["#FEB1A9", "#FEF2CD", "#CFF2B8", "#A6D8DB", "#BD9CF9"]
filtro_rubik.filtro('Einstein.jpg', 10, paleta_personalizada)
filtro_rubik.show()
```

También se puede guardar el último filtro aplicado, para ello se usa el método guardar. Si como parámetro no se pasa ninguna cadena de texto, se guardará con el nombre "imagen_modificada.jpg".

```python
filtro_rubik.guardar("Einstein_filtro_pastel.jpg") #se guarda la imagen

#mostrar imagen guardada
import skimage.io
skimage.io.imshow("Einstein_filtro_pastel.jpg")
```

También se puede usar el objeto para obtener la paleta de colores de una imagen dada. Para ello se usa el método obtener_paleta. Regresa una lista con los colores en formato hexadecimal.

```python
filtro_rubik.obtener_paleta("Einstein.jpg", 6) #imagen y número de colores
```

También se puede imprimir la paleta de colores con el método plot_paleta.

```python
filtro_rubik.plot_paleta("Einstein.jpg", 6) #imagen y número de colores
```

Por último, se puede transferir una paleta de colores de una imagen a otra. Para ello se usa el método transferir_paleta.
Se pasa como argumentos, primerp la imagen a modificar, la imagen a la que se le extrae la paleta de colores, el tamaño del cuadrado pixelado y el número de colores a extraer de la paleta de colores.

```python
filtro_rubik.transferir_paleta("Einstein.jpg", "artspring.png", cuadro_size=5, len_paleta=7)
filtro_rubik.show()
filtro_rubik.guardar('Einstein_artspring.jpg')

# Se muestra la imagen de la cual se transfirió la paleta de colores
skimage.io.imshow('artspring.jpg') 
```
